```python
import os
import rioxarray
import xarray as xr
import pandas as pd
import numpy as np
from rioxarray.merge import merge_arrays
from matplotlib import pyplot as plt
```

```python
lhi_list = []
for root, dirs, files in os.walk(r'E:\DATA_44842\NSW Government - Spatial Services\DEM\1 Metre'):
    for file in files:
        if '.asc' in file:
            lhi_list.append(rioxarray.open_rasterio(os.path.join(root,file)))
```

```python
merged = merge_arrays(lhi_list)
```

```python
merged.nbytes/1e6
```

```python
merged.rio.resolution()
```

```python
downsampled = merged.rio.reproject(merged.rio.crs, resolution=10)
downsampled.plot()
```

```python
ax = downsampled.plot(vmin=0)
ax.axes.set_aspect('equal')
```
